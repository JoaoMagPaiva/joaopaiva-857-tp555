```python
#DECLARAÇÃO DE BIBLIOTECAS

import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import pairwise_distances_argmin
from sklearn.metrics import silhouette_score
```

```python
#GERAÇÃO DE DADOS

N = 1000
# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close together.
X, y = make_blobs(n_samples=N, n_features=2, centers=4, cluster_std=1, center_box=(-10.0, 10.0), shuffle=True, random_state=1)
```

```python
#PLOT GRAFICO DOS DADOS GERADOS

idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)
idx2 = np.argwhere(y == 2)
idx3 = np.argwhere(y == 3)

fig = plt.figure(figsize=(7,7))
plt.plot(X[idx0,0], X[idx0,1], 'yx')
plt.plot(X[idx1,0], X[idx1,1], 'r*')
plt.plot(X[idx2,0], X[idx2,1], 'bo')
plt.plot(X[idx3,0], X[idx3,1], 'g+')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```

```python
#MÉTODO DO COTOVELO

sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(X)
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
#CRIAÇÃO DO MODELO UTILIZANDO N_CLUSTERS = 4

km = KMeans(n_clusters=4, random_state = 42)

km.fit(X)

#PLOT CENTROIDES

for i in range(0,4):
    print('Centroid %d: (%1.2f, %1.2f)' % (i,km.cluster_centers_[i][0],km.cluster_centers_[i][1]))
    
#NÚMERO DE ITERAÇÕES

print('Número de iterações: ' ,km.n_iter_)
```

```python
#FIGURA COM OS CLUSTERS E PONTOS

k_means_cluster_centers = km.cluster_centers_

k_means_labels = pairwise_distances_argmin(X, k_means_cluster_centers)

colors = ['#e83e08','#51e808', '#082fe8', '#e808a5']

for k, col in zip(range(4), colors):
    my_members = k_means_labels == k
    cluster_center = k_means_cluster_centers[k]
    plt.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='o', markersize = 6)
    plt.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=10)
```

```python
for n_cluster in range(2, 11):
    kmeans = KMeans(n_clusters=n_cluster).fit(X)
    label = kmeans.labels_
    coeficiente = silhouette_score(X, label, metric='euclidean')
    print("Para número de clusters={}, O coeficiente de silhueta é {}".format(n_cluster, coeficiente))
```

```python
#CRIAÇÃO DO MODELO UTILIZANDO N_CLUSTERS = 2

km = KMeans(n_clusters=2, random_state = 42)

km.fit(X)

#PLOT CENTROIDES

for i in range(0,2):
    print('Centroid %d: (%1.2f, %1.2f)' % (i,km.cluster_centers_[i][0],km.cluster_centers_[i][1]))
    
#NÚMERO DE ITERAÇÕES

print('Número de iterações: ' ,km.n_iter_)
```

```python
#FIGURA COM OS CLUSTERS E PONTOS

k_means_cluster_centers = km.cluster_centers_

k_means_labels = pairwise_distances_argmin(X, k_means_cluster_centers)

colors = ['#e83e08','#51e808']

for k, col in zip(range(2), colors):
    my_members = k_means_labels == k
    cluster_center = k_means_cluster_centers[k]
    plt.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='o', markersize = 6)
    plt.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=10)
```
