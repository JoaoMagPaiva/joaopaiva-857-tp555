```python
#DECLARAÇÃO DE BIBLIOTECAS

import matplotlib.pyplot as plt
from sklearn.datasets import fetch_20newsgroups
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.metrics import classification_report, precision_score, recall_score 
from sklearn.metrics import f1_score, roc_auc_score, precision_recall_curve
from sklearn.preprocessing import label_binarize
```

```python
#DEFININDO CATEGORIAS

categorias = ['talk.religion.misc', 'soc.religion.christian', 'sci.space', 'comp.graphics']
```

```python
#FAZENDO DOWNLOAD DE SET DE TREINO E VALIDACAO

treino = fetch_20newsgroups(subset='train', categories=categorias)
validacao = fetch_20newsgroups(subset='test', categories=categorias)
```

```python
#CRIAR O MODELO

vetor = CountVectorizer() #cria a matriz que conta o numero de vezes que cada termo aparece no texto

tipo_modelo = MultinomialNB() #define o classificador

model = make_pipeline(vetor, tipo_modelo) #cria modelo
```

```python
#TREINO E VALIDAÇÃO

model.fit(treino.data, treino.target) #treino do modelo

labels = model.predict(validacao.data) #teste de validacao
```

```python
#MATRIZ DE CONFUSÃO

fig = plt.figure(figsize=(7,7))
mat = confusion_matrix(validacao.target, labels)
sns.set(font_scale=1.8)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=treino.target_names, yticklabels=treino.target_names, cmap="Blues")
plt.xlabel('True label', fontsize=24)
plt.ylabel('Predicted label', fontsize=24)
plt.show()
```

```python
#CURVA ROC

# Getting the probabilities for each class.
probabilidades = model.predict_proba(validacao.data)

# Binarize the test targets.
binarizacao = label_binarize(validacao.target, classes=[0, 1, 2, 3])

# Compute ROC curve and ROC area for each class.
fpr = dict()
tpr = dict()
roc_auc = dict()
plt.figure(figsize=(30, 6))
lw = 2
for i in range(4):
    # Calculating ROC curve and ROC AUC.
    fpr[i], tpr[i], _ = roc_curve(binarizacao[:, i], probabilidades[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    # Plotting ROC curve for class i.
    ax = plt.subplot(1, 4, i + 1)
    plt.plot(fpr[i], tpr[i], color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC - Class '+str(i))
    plt.legend(loc="lower right")

plt.savefig('roc.png', dpi=600)
plt.show()
```

```python
#ACURÁCIA

print('Accuracy of Logistic Regression model is:', accuracy_score(validacao.target, labels)*100)
```

```python
#PRECISÃO

print('Precision score is:', precision_score(validacao.target, labels, average=None))
```

```python
#RECALL

print('Recall is:', recall_score(validacao.target, labels, average=None))
```

```python
#SCORE F COM PONDERAÇÃO 1

print('F1-score is:', f1_score(validacao.target, labels, average=None))
```

```python
#CURVA PRECISÃO X RECALL

precision = dict()
recall = dict()
for i in range(4):
    precision[i], recall[i], _ = precision_recall_curve(binarizacao[:, i], probabilidades[:, i])
    plt.plot(recall[i], precision[i], lw=2, label='class {}'.format(i))

plt.xlabel("recall")
plt.ylabel("precision")
plt.legend(loc="best")
plt.title("precision vs. recall curve")
plt.savefig('precision_recall_curve.png', dpi=600)
plt.show()
```
