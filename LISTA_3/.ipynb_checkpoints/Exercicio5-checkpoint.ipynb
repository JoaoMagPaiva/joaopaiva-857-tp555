```python
#DECLARACAO DE VARIAVEIS

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import LeavePOut
import csv
import pandas as pd
```

```python
df = pd.read_csv('./reg_poli.csv', header=None)

x = df[0].to_numpy()
y = df[1].to_numpy()
           
x = x.reshape(len(x),1)
y = y.reshape(len(x),1)

plt.plot(x,y,'r+')
plt.grid()

x_t, x_v, y_t, y_v = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#VALIDACAO HOLDOUT

erro_t = []
erro_v = []


for grau in range(1, 13):

    poly_features = PolynomialFeatures(degree=grau, include_bias=False) # cria o polinomio

    std_scaler = StandardScaler() # gera o escalonamento de features por padronizacao

    lin_reg = LinearRegression() # gera o regressor linear

    polynomial_regression = Pipeline([
        ("poly_features", poly_features),
        ("std_scaler", std_scaler),
        ("lin_reg", lin_reg),
    ]) # cria o modelo
    
    polynomial_regression.fit(x_t, y_t) # treina o modelo
    
    y_t_predicao = polynomial_regression.predict(x_t)
    
    y_v_predicao = polynomial_regression.predict(x_v)    
    
    erro_t.append(np.sqrt(mean_squared_error(y_t, y_t_predicao)))

    erro_v.append(np.sqrt(mean_squared_error(y_v, y_v_predicao)))

# Plot results.
plt.figure(figsize=(10, 10))
plt.plot(range(1, 13), erro_t, label='Treinamento')
plt.plot(range(1, 13), erro_v, label='Validação')
plt.xlabel('Ordem do polinômio',  fontsize=14)
plt.ylabel('Erro quadrático médio',  fontsize=14)
plt.xticks(range(0, 13, 2))
plt.xlim([1, 12])
plt.legend()
plt.grid()

#Show the plot.
plt.show()

plt.savefig('ERROS TREINAMENTO E VALIDACAO x GRAU DO POLINOMIO EXERCICIO 5 HOLDOUT', dpi = 600)
```

```python
#VALIDAÇÃO K-FOLD

kfold = KFold(n_splits=10, shuffle=True, random_state=42) # gerando os folders

erro_t = []
erro_v = []
vetor_media_erro = []
desvio_padrao = []

for grau in range(1, 13):

    poly_features = PolynomialFeatures(degree=grau, include_bias=False) # cria o polinomio
    
    std_scaler = StandardScaler() # gera o escalonamento de features por padronização

    lin_reg = LinearRegression() # gera o regressor linear

    polynomial_regression = Pipeline([
        ("poly_features", poly_features),
        ("std_scaler", std_scaler),
        ("lin_reg", lin_reg),
    ]) # cria o modelo

    lin_scores = cross_val_score(polynomial_regression, x, y, scoring='neg_mean_squared_error', cv=kfold)
    scores = np.sqrt(-lin_scores)
    vetor_media_erro.append(scores.mean())
    desvio_padrao.append(scores.std())

# Plot results.
plt.figure(figsize=(10, 5))
ax = plt.subplot(1, 2, 1)
plt.plot(range(1, 13), vetor_media_erro)
plt.xlabel('Ordem do polinômio',  fontsize=14)
plt.ylabel('Média do erro quadrático médio',  fontsize=14)
plt.xticks(range(0, 13, 2))
plt.xlim([1, 12])
plt.grid()

ax = plt.subplot(1, 2, 2)
plt.plot(range(1, 13), desvio_padrao)
plt.xlabel('Ordem do polinômio',  fontsize=14)
plt.ylabel('Desvio padrão',  fontsize=14)
plt.xticks(range(0, 13, 2))
plt.xlim([1, 12])
plt.grid()

#Show the plot.
plt.show()

plt.savefig('ERROS TREINAMENTO E VALIDACAO x GRAU DO POLINOMIO EXERCICIO 5 K-FOLD', dpi = 600)
```

```python
#VALIDACAO LEAVE-P-OUT

loocv = LeaveOneOut()

vetor_media_erro = []
desvio_padrao = []

for grau in range(1, 13):

    poly_features = PolynomialFeatures(degree=grau, include_bias=False)

    std_scaler = StandardScaler()

    lin_reg = LinearRegression()

    polynomial_regression = Pipeline([
        ("poly_features", poly_features),
        ("std_scaler", std_scaler),
        ("lin_reg", lin_reg),
    ])

    lin_scores = cross_val_score(polynomial_regression, x, y, scoring='neg_mean_squared_error', cv=loocv)
        
    scores = np.sqrt(-lin_scores)
    vetor_media_erro.append(scores.mean())
    desvio_padrao.append(scores.std())
    
# Plot results.
plt.figure(figsize=(10, 5))
ax = plt.subplot(1, 2, 1)
plt.plot(range(1, 13), vetor_media_erro)
plt.xlabel('Ordem do polinômio', fontsize=14)
plt.ylabel('Média do erro quadrático Médio', fontsize=14)
plt.xticks(range(0, 13, 2))
plt.xlim([1, 12])
plt.grid()

ax = plt.subplot(1, 2, 2)
plt.plot(range(1, 13), desvio_padrao)
plt.xlabel('Ordem do polinômio', fontsize=14)
plt.ylabel('Desvio padrão', fontsize=14)
plt.xticks(range(0, 13, 2))
plt.xlim([1, 12])
plt.grid()

#Show the plot.
plt.show()

plt.savefig('ERROS TREINAMENTO E VALIDACAO x GRAU DO POLINOMIO LEAVE-P-OUT', dpi = 600)
```

```python
#CURVAS DE APRENDIZADO

graus = [1, 3, 6, 9, 12]

plt.figure(figsize=(40, 10))

for i in range(len(graus)):
    ax = plt.subplot(1, len(graus), i + 1)

    poly_features = PolynomialFeatures(degree=graus[i], include_bias=False)
    
    std_scaler = StandardScaler()
    
    lin_reg = LinearRegression()

    polynomial_regression = Pipeline([
        ("poly_features", poly_features),
        ("std_scaler", std_scaler),
        ("lin_reg", lin_reg),
    ])

    erro_t = []
    erro_v = []
    
    for m in range(1,len(x_t)+1):

        # Perform polynomial regression.
        polynomial_regression.fit(x_t[:m], y_t[:m])

        # Use the trained model for prediction of the training set.
        y_t_predicao = polynomial_regression.predict(x_t[:m])

        # Use the trained model for prediction of the validation set.
        y_v_predicao = polynomial_regression.predict(x_v)       

        # Calculate MSE for training set.
        erro_t.append(mean_squared_error(y_t[:m], y_t_predicao))

        # Calculate MSE for validation set.
        erro_v.append(mean_squared_error(y_v, y_v_predicao))     

    plt.plot(range(1,len(x_t)+1), np.sqrt(erro_t), label='Conjunto de treinamento')
    plt.plot(range(1,len(x_t)+1), np.sqrt(erro_v), label='Conjunto de validação')
    plt.yscale('log')
    plt.xlabel('Tamanho do conjunto de treinamento', fontsize=14)
    plt.ylabel('Raiz Quadrada do Erro Quadrático Médio', fontsize=14)
    plt.grid()
    plt.title('Polinômio de ordem '+str(graus[i]))
    plt.legend(loc='best')
    
#Show the plot.
plt.show()
```

```python
#TREINAMENTO DO MODELO PARA UM POLINOMIO DE GRAU 6

poly_features = PolynomialFeatures(degree=6, include_bias=False)

std_scaler = StandardScaler()

lin_reg = LinearRegression()

model = Pipeline([
    ("poly_features", poly_features),
    ("std_scaler", std_scaler),
    ("lin_reg", lin_reg),
])

model.fit(x, y)

y_pred = model.predict(x)
```

```python
#PLOT COMPARACAO HIPOTESE DE GRAU 6 COM DADOS RUIDOSOS
fig = plt.figure(figsize=(7,7))
plt.plot(x, y, 'x', label='Dado ruidoso')
plt.plot(x, y_pred, 'x', label='Hipótese')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.legend()
plt.grid()
#Show the plot.
plt.show()

plt.savefig('COMPARACAO DADOS RUIDOSOS COM A HIPOTESE DE GRAU 6', dpi = 600)
```
