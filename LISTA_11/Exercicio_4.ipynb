```python
#DECLARAÇÃO DE BIBLIOTECAS

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn.metrics import mean_squared_error
```

LÓGICA E

```python
#GERAÇÃO DE DADOS

N = 1000

x1 = np.random.randint(0,2,N)
x2 = np.random.randint(0,2,N)

y = x1 & x2

x1 = x1 + 0.1*np.random.randn(N,)

x2 = x2 + 0.1*np.random.randn(N,)

x0 = np.ones((N,))
X = np.c_[x0,x1,x2,y]
```

```python
#PLOT DOS DADOS

idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)

fig = plt.figure(figsize=(5,5))
plt.plot(x1[idx0], x2[idx0], '.', label='Class 0')
plt.plot(x1[idx1], x2[idx1], 'rx', label='Class 1')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.title('Nosy AND function')
plt.legend()
plt.savefig('noisy_and_gate.png', dpi=600)
plt.show()
```

```python
#TREINAMENTO

epocas = 2000

alpha = 0.1

iteracao = 0
#criando estruturas vazias

erro = np.zeros(epocas)
pesos = np.zeros((1,3))
pesos_otimos = np.zeros((1,3))
erro_minimo = 1000

for epoca in range(epocas):
   
    np.random.permutation(X)
   
    for iteracao in range(N):
        
        y_calc = pesos[0,0]*X[iteracao,0] + pesos[0,1]*X[iteracao,1] + pesos[0,2]*X[iteracao,2]
        erro[epoca] = X[iteracao,3] - y_calc
   
        if(erro[epoca] < erro_minimo):
            pesos_otimos[0,0] = pesos[0,0]
            pesos_otimos[0,1] = pesos[0,1]
            pesos_otimos[0,2] = pesos[0,2]
        
        pesos[0,0] = pesos[0,0] + (alpha*erro[epoca]*X[iteracao,0])
        pesos[0,1] = pesos[0,1] + (alpha*erro[epoca]*X[iteracao,1])
        pesos[0,2] = pesos[0,2] + (alpha*erro[epoca]*X[iteracao,2])
        
print(pesos_otimos)
```

```python
#PLOT ERRO POR EPOCA

fig = plt.figure(figsize=(5,5))
plt.plot(np.linspace(1,2000,2000), erro)
plt.xlabel('Epoca', fontsize=14)
plt.ylabel('Erro', fontsize=14)
plt.title('Erro por Epoca')
plt.show()
```
