```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

LeakyReLU (UTILIZAREI UM ALPHA DE 0.2 QUE, SEGUNDO A REFERÃŠNCIA POSSUI NORMALMENTE BOA PERFORMANCE)

```python
def leakyrelu(z):
    
    ans = np.maximum(0, z)
    
    for i in range(len(ans)):
        
        if (z[i] <= 0):
            
            ans[i] = 0.2*(z[i])
    
    return ans

def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)
```

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, leakyrelu(z), "m-.", linewidth=2, label="LeakyReLU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Activation function", fontsize=14)
plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(z, derivative(leakyrelu, z), "m-.", linewidth=2, label="LeakyReLU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -0.2, 1.2])

plt.show()
```

ELU (UTILIZAREI UM ALPHA DE 1 QUE FOI DITO PELA REFERÃŠNCIA COMO COMUM

```python
def elu(z):
    
    ans = np.maximum(0, z)
    
    for i in range(len(ans)):
        
        if (z[i] <= 0):
            
            ans[i] = 0.1 * (np.exp(z[i]) - 1)
    
    return ans
```

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, elu(z), "m-.", linewidth=2, label="ELU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Activation function", fontsize=14)
plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(z, derivative(elu, z), "m-.", linewidth=2, label="ELU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -0.2, 1.2])

plt.show()
```

BATCH NORMALIZATION GAMMA 0.1 E BETA 1 (NÃ£o entendi a derivada zerar)

```python
def batchnorm_forward(z, gamma=0.1, beta=0, eps=0.000001):
    
    sample_mean = z.mean(axis=0)
    sample_var = z.var(axis=0)
    
    std = np.sqrt(sample_var + eps)
    z_centered = z - sample_mean
    z_norm = z_centered / std
    out = gamma * z_norm + beta
    

    return out
```

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, batchnorm_forward(z), "m-.", linewidth=2, label="BatchNormalization")
plt.grid(True)
plt.legend(loc="upper right", fontsize=14)
plt.title("Activation function", fontsize=14)
plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(z, derivative(batchnorm_forward, z), "m-.", linewidth=2, label="LeakReLU")
plt.grid(True)
plt.legend(loc="upper right", fontsize=14)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -0.2, 1.2])

plt.show()
```
