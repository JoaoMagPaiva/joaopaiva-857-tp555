```python
#DECLARAÇÃO DE BIBLIOTECAS

import numpy as np
from imageio import imread
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.utils import shuffle
```

```python
#LEITURA DA IMAGEM

image = imread("img2.jpg")

print('Dimensôes da Matrix "image":', np.shape(image))
```

```python
#PLOT IMAGEM ORIGINAL

plt.figure()
plt.imshow(image)
```

```python
#SEPARAÇÃO EM DADOS ALTURA, LARGURA E DIMENSÕES

w, h, d = original_shape = tuple(image.shape)
print("Largura:",w)
print("Altura:",h)
print("Dimensão:",d)

#CONVERSÃO EM UMA MATRIX 2-D
image = np.array(image, dtype=np.float64) / 255
image_array = np.reshape(image, (w * h, d))
print('')
print('Separação 2d: ')
print(image_array)
print('Dimensões da nova matrix 2d: ', image_array.shape)
```

2 CLUSTERS

```python
#TREINAMENTO K-MEANS

#Tomando 1000 pontos aleatórios da imagem
image_amostras_2 = shuffle(image_array)[:1000]

#criando o modelo
km_2 = KMeans(n_clusters=2, random_state = 42)

km_2.fit(image_amostras_2)

#predição com a matrix 2d
pred_2 = km_2.predict(image_array)

for i in range(0,2):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,km_2.cluster_centers_[i][0],km_2.cluster_centers_[i][1],km_2.cluster_centers_[i][2]))
```

4 CLUSTERS

```python
#TREINAMENTO K-MEANS

#Tomando 1000 pontos aleatórios da imagem
image_amostras_4 = shuffle(image_array)[:1000]

#criando o modelo
km_4 = KMeans(n_clusters=4, random_state = 42)

km_4.fit(image_amostras_4)

#predição com a matrix 2d
pred_4 = km_4.predict(image_array)

for i in range(0,4):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,km_4.cluster_centers_[i][0],km_4.cluster_centers_[i][1],km_4.cluster_centers_[i][2]))
```

```python
#TREINAMENTO K-MEANS

#Tomando 1000 pontos aleatórios da imagem
image_amostras_16 = shuffle(image_array)[:1000]

#criando o modelo
km_16 = KMeans(n_clusters=16, random_state = 42)

km_16.fit(image_amostras_16)

#predição com a matrix 2d
pred_16 = km_16.predict(image_array)

for i in range(0,16):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,km_16.cluster_centers_[i][0],km_16.cluster_centers_[i][1],km_16.cluster_centers_[i][2]))
```

```python
#FUNÇÃO DE RECRIAR A IMAGEM

def recreate_image(codebook, labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image
```

```python
#RECRIANDO A IMAGEM PARA 2 CORES

plt.figure(1)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(km_2.cluster_centers_, pred_2, w, h))
plt.title('2 CORES')

#RECRIANDO A IMAGEM PARA 4 CORES
plt.figure(2)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(km_4.cluster_centers_, pred_4, w, h))
plt.title('4 CORES')

#RECRIANDO A IMAGEM PARA 8 CORES
plt.figure(3)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(km_8.cluster_centers_, pred_8, w, h))
plt.title('8 CORES')

#RECRIANDO A IMAGEM PARA 16 CORES
plt.figure(4)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(km_16.cluster_centers_, pred_16, w, h))
plt.title('16 CORES')
```
